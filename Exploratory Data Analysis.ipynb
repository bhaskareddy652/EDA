```python
columns= ['LB', 'AC', 'DS', 'DP', 'MLTV', 'Width', 'Tendency', 'NSP']
mean=df[columns].mean()
df[columns]=df[columns].fillna(mean)
```

```python
columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']
outliers_df = pd.DataFrame()
for col in columns:
    Q1 = df[col].quantile(0.25) 
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_df[col] = outliers[col]
print(f"Outliers in {col}:")
print(outliers)
print('-' * 50)
```

```python
df_new = df[~df.isin(outliers_df).any(axis=1)]
print("Dataframe without outliers:")
display(df_new)
```

Violin Plot

```python

num_cols_subset = df_new.select_dtypes(include=np.number).columns[0:14]
# columns and rows needed for the subplot grid
num_plots = len(num_cols_subset)
num_rows = int(np.ceil(num_plots / 4))  # we can Adjust the number of columns per row as needed
# it hold a multiple subplots
plt.figure(figsize=(16, num_rows * 4))  #we can  Adjust height based on number of rows
# here im ploting a violin plot for each numerical column in the subset
for i, column in enumerate(num_cols_subset):
    plt.subplot(num_rows, 4, i + 1)  #  we Adjust subplot grid size (rows, cols)
    sns.violinplot(x=df_new[column])
    plt.title(f'Violin Plot of {column}')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

Distribution Shape: The width of the violin plot at different values indicates the density of the data at those values. Wider sections represent higher density, meaning more data points are concentrated there:

1.
Central Tendency: The white dot in the center of the plot shows the median of the data. The thick black bar in the center represents the interquartile range (IQR), which is the middle 50% of the data
2.
Symmetry: The plot is mirrored, so you can easily see if the data distribution is symmetric or skewed. If one side is wider than the other, it indicates skewness.

```python
correlation_matrix = df_new.corr()
print("Correlation Matrix:")
display(correlation_matrix)
# Finding highly correlated variables
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
  for j in range(i + 1, len(correlation_matrix.columns)):
    if abs(correlation_matrix.iloc[i, j]) > 0.7:  # Threshold for high correlation
      high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))
print("\nHighly Correlated Variables:")
for pair in high_corr_pairs:
  print(f"{pair[0]} and {pair[1]} - Correlation: {pair[2]:.2f}")
# Finding less correlated variables
low_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
  for j in range(i + 1, len(correlation_matrix.columns)):
    if abs(correlation_matrix.iloc[i, j]) < 0.3:  # Threshold for low correlation
      low_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))
print("\nLess Correlated Variables:")
for pair in low_corr_pairs:
  print(f"{pair[0]} and {pair[1]} - Correlation: {pair[2]:.2f}")
```

Analysis Of correlated variables:

Highly correlated variables indicate a strong linear relationship between two variables
This can be useful for understanding the underlying relationships in the data and for making predictions.
For example, if two variables are highly correlated, it might be possible to use one variable to predict the other.
However, it is important to note that correlation does not imply causation.
Just because two variables are highly correlated, it does not mean that one variable causes the other.

Less correlated variables indicate a weak linear relationship between two variables.
This means that the variables are not strongly related to each other.
This can be useful for identifying variables that are independent of each other.
For example, if two variables are less correlated, it might be possible to use them as independent predictors in a model.
